# Clustering

```python
from sklearn.preprocessing import scale
from sklearn.manifold import TSNE 
from sklearn.cluster import KMeans
```

```python
data={}
X_df=df[numeric]
for col in categorical:
    data[col]=pd.get_dummies(df[col])
    X_df=pd.concat([X_df,data[col]],axis=1)
X_df.head()
    

```

```python
X=scale(X_df)
X
```

```python
tsne=TSNE(verbose=1,random_state=20,perplexity=100)
X_embadded=tsne.fit_transform(X)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(X_embadded[:,0],X_embadded[:,1])
plt.title("TSNE with no labels",fontweight='bold',size=10,fontfamily='monospace')
plt.show()
```

```python
cluster=KMeans(n_clusters=12)
cluster.fit(X)
y_pred=cluster.predict(X)
```

```python
palette=sns.hls_palette(12, l=.2, s=.9)
plt.figure(figsize=(10,5))
sns.scatterplot(X_embadded[ : ,0],X_embadded[ : ,1],hue=y_pred,palette=palette)
plt.title("t-SNE plot",size=10,fontweight='bold',fontfamily='monospace')
plt.show()
```

```python
df['cluster']=y_pred
df.head()
```

```python
df.groupby('cluster')["math score",'reading score','writing score'].mean().style.background_gradient('Reds')
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(data=df,x='cluster',y='writing score')
plt.title("Strip plot for writing Score",fontweight='bold',fontfamily='monospace',size=10)
plt.xlabel('CLUSTER',fontweight='light',fontfamily='monospace',size=10)
plt.ylabel('Writing Score',fontweight='light',fontfamily='monospace',size=10)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(data=df,x='cluster',y='reading score')
plt.title("Strip plot for reading score",size=10,fontweight='bold',fontfamily='monospace')
plt.xlabel('CLUSTER',size=10,fontfamily='monospace',fontweight='light')
plt.ylabel('Reading Score',size=10,fontfamily='monospace',fontweight='light')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.stripplot(data=df,x='cluster',y='math score')
plt.title("Strip plot for Math Score",size=10,fontweight='bold',fontfamily='monospace')
plt.xlabel("Cluster",size=10,fontweight='light',fontfamily='monospace')
plt.ylabel("Math Score",size=10)
plt.show()
```
